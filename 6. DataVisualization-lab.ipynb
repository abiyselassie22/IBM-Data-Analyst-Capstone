# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""

# The read_sql_query runs the sql query & returns the data as a dataframe
df=pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt
df.hist(bins=10, color='r')
plt.title("Histogram of Annual USD Salaries for ConvertedComp")
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.show()
```

```python
# Alternative to plot a histogram of ConvertedComp
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_sql_query("SELECT*FROM master",conn)
plt.hist(df['ConvertedComp'], color='r', bins= 10)
plt.title("Histogram of Annual USD Salaries for ConvertedComp")
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# The read_sql_query runs the sql query & returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color='b')
plt.title('Correlation between Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
# Normalizing the Age attribute to determine the bubble size for the Age column
norm_age=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())

# The read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s=norm_age*1000, alpha=0.5, figsize=(8,6), color='b')
plt.title('Relationship of WorkWeekHrs & CodRevHrs based on varying Age')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# The table we desire to describe
table_name = 'DatabaseDesireNextYear'

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df=pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
# SQL query to fetch all data from the table named 'DatabaseDesireNextYear'
QUERY = """
SELECT * FROM DatabaseDesireNextYear;
"""

# Execute the query & fetch data into a Dataframe
table_data_df1 = pd.read_sql_query(QUERY,conn)

# Print the Dataframe containing data from the 'DatabaseDesireNextYear'table
print(table_data_df1)
```

```python
# your code goes here
QUERY = """ 
SELECT COUNT(*) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
Group by DatabaseDesireNextYear
Order by Count
DESC LIMIT 5
"""

# Loading the data from the table into a dataframe using read_sql_query
DatabaseDesireNextYear=pd.read_sql_query(QUERY,conn)

labels=DatabaseDesireNextYear['DatabaseDesireNextYear']
color_list=['lightgreen','red','blue','orange','skyblue']

plt.figure(figsize=(10,6))
DatabaseDesireNextYear['Count'].plot(kind='pie', autopct='%1.1f%%',startangle=90,shadow=True,labels=labels,pctdistance=0.5,colors=color_list)
plt.title('Top 5 Databases that Respondents wish to learn Next Year')
plt.axis('equal')
plt.show()
```

```python
# SQL query to fetch all data from the table named 'DatabaseWorkedWith'
QUERY = """
SELECT * FROM DatabaseWorkedWith;
"""

# Execute the query and fetch data into a DataFrame
table_data_df2 = pd.read_sql_query(QUERY, conn)

# Print the DataFrame containing data from the 'DatabaseWorkedWith' table
print(table_data_df2)
```

```python
# SQL query to fetch only MySQL data that Respondents indicated to work with
QUERY = """
SELECT DatabaseWorkedWith, Count(Distinct Respondent) as Count
FROM DatabaseWorkedWith
Group by Respondent
Having Count(DatabaseWorkedWith)=1 and DatabaseWorkedWith='MySQL';
"""

# Execute the query and fetch data into a DataFrame
table_data_df2 = pd.read_sql_query(QUERY, conn)

# Print the DataFrame containing data from the 'DatabaseWorkedWith' table
print(table_data_df2)
```

```python
# SQL query to fetch all data from the table named 'LanguageDesireNextYear'
QUERY = """
SELECT * FROM LanguageDesireNextYear;
"""

# Execute the query and fetch data into a DataFrame
table_data_df3 = pd.read_sql_query(QUERY, conn)

# Print the DataFrame containing data from the 'LanguageDesireNextYear' table
print(table_data_df3)
```

```python
# SQL query to fetch the most popular language that Respondents wish to learn next year.
QUERY = """
SELECT COUNT(*) as Count, LanguageDesireNextYear,
Rank() Over (Order by Count(LanguageDesireNextYear) DESC) As Language_Rank
FROM LanguageDesireNextYear
WHERE LanguageDesireNextYear IS NOT NULL
Group by LanguageDesireNextYear
Order by Count
DESC LIMIT 5;
"""
# Execute the query and fetch data into a DataFrame
table_data_df3 = pd.read_sql_query(QUERY, conn)

# Print the DataFrame containing data from the 'LanguageDesireNextYear' table
print(table_data_df3)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age>= 30 AND Age<=35
"""

# Loading the data from the table into a dataframe using read_sql_query
df=pd.read_sql_query(QUERY,conn)

# Group by the age and take the median for WorkWeekHrs and CodeRevHrs
grouped_df=df.groupby('Age').median(numeric_only=True)

# Create a stacked bar chart & customize it
grouped_df.plot(kind='bar', stacked=True, color=['red','blue'], label=['WorkWeekHrs', 'CodeRevHrs'])

plt.title('Stacked Bar chart of Median WorkWeekHrs and CodeRevHrs for Age group 30-35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.xticks(rotation=45)
plt.legend(title='Columns', loc='lower right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
# Loading the data from the table into a dataframe using read_sql_query
df=pd.read_sql_query(QUERY,conn)

# Group by the age group and take the median for ConvertedComp
grouped_df2=df.groupby('Age').median(numeric_only=True)

# Create a Line chart & customize it
grouped_df2.plot(kind='line', color='r', label='ConvertedComp')

plt.title('Line chart of Median ConvertedComp for all Age group 45-60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.legend(loc='upper left')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY="""
SELECT MainBranch, COUNT(*) as Count
FROM master
Group by MainBranch
"""

# Loading the data from the table into a dataframe using read_sql_query
df3=pd.read_sql_query(QUERY,conn)

# Set the MainBranch column as the index
df3.set_index('MainBranch', inplace=True)

# Create a Horizontal Bar chart using the Column MainBranch
ax=df3['Count'].plot(kind='barh', figsize=(8,5), color='blue')
plt.title('Horizontal Bar Chart of the Column MainBranch')
plt.xlabel('Count')

# Annotate value labels to each Count of the MainBranch
for index, value in enumerate(df3['Count']):
    label = format(int(value), ',') # Format int with commas 
    plt.annotate(label, xy=(value + 5, index - 0.10), color='red')

plt.show()
```
